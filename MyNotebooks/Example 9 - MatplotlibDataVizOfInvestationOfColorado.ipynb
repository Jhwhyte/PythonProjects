```python
#imports the libaries needed to both display the data and randomly generate points
import matplotlib.pyplot as plt
import random

#this parameterized the graph so that it comes out as a rectangle, set up in a x-y fashion #using lists
plt.plot( [-109.05,-102.05,-102.05,-109.05,-109.05], [41,41,37,37,41] )

#assings the x-axis with the name longitude
plt.xlabel('Longitude')
#assings the y-axis with the name longitude
plt.ylabel('Latitude')
#assings the title to Colorado
plt.title('Colorado')

#used later to create blue dots
sym = 'bo'

#loop and generate 500 plots
for points in range(0, 500):
    #randomly and uniformly generate plots so long as it's within bounds for these lats and #lngs
    rand_lat = random.uniform(-109.05, -102.05)
    rand_lng = random.uniform(37, 41)
#generates plots parameterized for the upper right corner
    if rand_lat > -105.5 and rand_lat <= -102.05 and rand_lng > 39 and rand_lng <= 41:
        plt.plot(rand_lat, rand_lng, sym)
#generates plots parameterized for the upper left corner
    elif rand_lat < -105.5 and rand_lat >= -109.05 and rand_lng > 39 and rand_lng <=41:
        plt.plot(rand_lat, rand_lng, '>')
#generates plots parameterized for the lower right corner
    elif rand_lat < -105.5 and rand_lat >= -109.05 and rand_lng > 37 and rand_lng <=39:
        plt.plot(rand_lat, rand_lng, '*')
#generates plots parameterized for the lower left corner
    else:
        plt.plot(rand_lat, rand_lng, 'x')
#function call to the library, used to display the graph         
plt.show()



#==========PART2===================


#Name: James Whyte
#Date: 3/16/2018
#Project title: Part1 and part2 of  lab03
#Description: This program plots random latitude and longitudes using the Matplotlib and random libraries in order to generate
#the plots and uses matplotlib to graphically display them. The first part is the general layout of a specific set of points and
#random points. The second part is for users to enter their own custom splits for the graph.


#import the modules to display and generate random plots
import matplotlib.pyplot as plt
import random


#creating a function in order for it to be called when the user inputs a split point. it checks whether those points fall within
#the set boundaries of the project
def validInput(floatVal, currInput):
    #I check if the variableI assigned for the firsr question matches the string latquad_user
    if currInput == 'latquad_user' :
        #if it does then checks if it's within the bounds
        if (floatVal >=-109.05 and floatVal <=-102.05):
            #if so, then set return true and carry on to the rest of the program
            return True
        else:
            #otherwise it will be false and start the loop over again
            return False
        #if it wasn't that variable it must be the other, so do the same checks for matching string and varaible
    elif currInput == 'lngquad_user':
        #compare the set boundaries
        if (floatVal >= 37 and floatVal <= 41):
            #if that is true returns true
            return True
        else:
            #otherwise false
            return False
    #if user didn't enter either then was the wrong input, so false
    else:
        return False



#Assuming the user input to be an integer, this asks and stores the user data for the number of sites
samplesite_user = input("Enter the number of sample sites: ")

#for use in the while-loop
boo = False

#starting a while loop in order to continue asking the question for latitude, if it wasn't valid
#to iterate the loop, I ahve an outer variable boolean that stays False
while(boo == False):
    #asks the user for their desired latitude stores it in a variable
    latquad_user = input("Enter the latitude of the quadrant split: ")
    #checks to see if that latitude is valid, if so then it's true and carry on
    if(validInput(latquad_user,'latquad_user') == True):
        #assign to true if the above was true
        boo = True
    else:
        #else assign to false and start over
        boo = False

#same as above, keep a boolean for the while loop
boo2 = False
#kick start the while and keep asking the user for longitude if it's out of bounds
while(boo2 == False):
    #asks user for longitude and stores it in a variable
    lngquad_user = input("Enter the longitude of the quadrant split: ")
    #if variable is valid in calling the method and passing in the parameters lngquad_user,
    #and 'lngquad_user, if true carry on
    if(validInput(lngquad_user,'lngquad_user') == True):
        #if true then to the next line
        boo2 = True
    else:
        #if false then start over
        boo2 = False


#this creates the create and creates boundaries for the upperright, lowerright, bottomleft, bottomright
#in a kind of list of x-y variables
plt.plot( [-109.05,-102.05,-102.05,-109.05,-109.05], [41,41,37,37,41] )

#labels the x-axis with the name Longitude
plt.xlabel('Longitude')
#labels the y-axis with the name latitude
plt.ylabel('Latitude')
#was not required but I titled it Colorado
plt.title('Colorado')

#To later use for the blue dots in the looping structure
sym = 'bo'

#for loops to get random plots for 0 to what the user asks for, for his/her samplesite number
for points in range(0, samplesite_user):
    #randomly and uniform form the lines of latitude, the upper and lower bounnds for latitude
    rand_lat = random.uniform(-109.05, -102.05)
    #randomly and uniform form the lines of longitude, the upper and lower bounnds for longitude
    rand_lng = random.uniform(37, 41)
    
    #if-else conditions to parameterize the grid so that specific points fall within specific quadrants
    #This if plots blue dots for the upper right corner
    if rand_lat > latquad_user and rand_lat <= -102.05 and rand_lng > 39 and rand_lng <= 41:
        plt.plot(rand_lat, rand_lng, sym)
    #This if plots points for the upper left corner
    elif rand_lat < latquad_user and rand_lat >= -109.05 and rand_lng > 39 and rand_lng <=41:
        plt.plot(rand_lat, rand_lng, '>')
    #this if plots points for the bottom left corner
    elif rand_lat < latquad_user and rand_lat >= -109.05 and rand_lng > lngquad_user and rand_lng <=39:
        plt.plot(rand_lat, rand_lng, '*')
    #this else plots for the bottom right corner
    else:
        plt.plot(rand_lat, rand_lng, 'x')


#method used to display the graph
plt.show()
```
