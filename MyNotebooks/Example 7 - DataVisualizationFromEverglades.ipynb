#### Purpose: Collected data from usgs and graphed observations on water level and climate measurements over the Everglades. The USGS has a network of ground observing stations across the Everglades. Data collected at these stations are available from https://sofia.usgs.gov/eden/stationlist.php Task was to select 3 stations, download data on water level, evapotranspiration, and rainfall, and plot the data for two yea(2017 and 2012).

#### By: James Whyte


```python
# importing the needed libraries to do spatial analysis
import pandas as pd
import matplotlib.pyplot as plt


# This line creates three rows and one col for the following graphs to be inserted into. Also they share the x-axis
fig, axes = plt.subplots(nrows=3, ncols=1, sharex=True)

# Readining in the csv data from my desktop. Skipping the first 4 rows to get to the data
df = pd.read_csv(r'C:\Users\new\Desktop\EDEN_7_2012\1541644891_rainfall.csv', skiprows=4)

# this gives the line plot a color, no legend, places it in the first row of 'fig' and gives it a title
ax2 = df.plot(color="green",kind='line', legend = False, ax=axes[0], figsize=(12,15), title='Eden_7 (2012)')

# names the y-axis
ax2.set_ylabel("Rainfall (inches)")

# creating even spacing the for x-ticks of the 365 days in a year
ax2.set_xticks([0,65,125,185,245,305,365])

# creating the array of months and setting the fontsize to large for those variables on the x-axis
ax2.set_xticklabels(['Jan','Mar','May','Jul','Sep','Nov','Jan'], fontsize = 'large')


# Making the y-ticks vertical as in the lab example giving it a smaller fontsize to adjust for sizes
plt.setp(ax2.get_yticklabels(), rotation='vertical', fontsize=12)




# Readining in the csv data from my desktop. Skipping the first 4 rows to get to the data
df = pd.read_csv(r'C:\Users\new\Desktop\EDEN_7_2012\1541644891_water_level.csv', skiprows=4)

# this gives the line plot a color, no legend, places it in the first second of 'fig' and gives it a title
ax3 = df.plot(color="orange",kind='line', legend = False, ax=axes[1])

# names the y-axis
ax3.set_ylabel("Water Level (ft above NADV88)")

# creating even spacing the for x-ticks of the 365 days in a year
ax3.set_xticks([0,65,125,185,245,305,365])

# creating the array of months and setting the fontsize to large for those variables on the x-axis
ax3.set_xticklabels(['Jan','Mar','May','Jul','Sep','Nov','Jan'], fontsize = 'large')

# Making the y-ticks vertical as in the lab example giving it a smaller fontsize to adjust for sizes
plt.setp(ax3.get_yticklabels(), rotation='vertical', fontsize=12)



# Readining in the csv data from my desktop. Skipping the first 4 rows to get to the data
df = pd.read_csv(r'C:\Users\new\Desktop\EDEN_7_2012\1541644891_et.csv', skiprows=4)

# this gives the line plot a color, no legend, places it in the first second of 'fig' and gives it a title
ax = df.plot(color="purple",kind='line', legend = False, ax=axes[2])

# names the y-axis
ax.set_ylabel("Evapotranspiration (inches)")

# creating even spacing the for x-ticks of the 365 days in a year
ax.set_xticks([0,65,125,185,245,305,365])

# creating the array of months and setting the fontsize to large for those variables on the x-axis
ax.set_xticklabels(['Jan','Mar','May','Jul','Sep','Nov','Jan'], fontsize = 'large')

# makes the y ticks vertical
plt.yticks(rotation=90)

# plotting those line graphs and joining them together with hspace = 0
fig.subplots_adjust(hspace=0)







```
